### ¿ Qué es una serie de tiempo?

Una serie de tiempo es una secuencia de datos medidos en momentos sucesivos, generalmente a intervalos regulares. Estas series se utilizan para analizar patrones, tendencias y comportamientos a lo largo del tiempo. Las series de tiempo son comunes en diversas disciplinas, como la economía, la meteorología, la ingeniería y la ciencia de datos.

### ¿ Que son los Modelos no estacionarios?

Los modelos no estacionarios son aquellos que describen series de tiempo cuyos estadísticos, como la media y la varianza, cambian a lo largo del tiempo. Estos modelos son útiles para analizar datos que presentan tendencias, estacionalidades o cambios estructurales. Algunos ejemplos de modelos no estacionarios incluyen el modelo de tendencia lineal, el modelo de tendencia exponencial y el modelo de estacionalidad.

### ¿Que es el box cox?

El Box-Cox es una transformación de datos utilizada para estabilizar la varianza y hacer que los datos sean más normales. Esta transformación se aplica a series de tiempo para mejorar la precisión de los modelos de pronóstico y que se acerque mas a una distribución normal. Tambien sirve para estabilizar la varianza (homocedasticidad).

### ¿Cual es el metodo de guerrero?

El método de Guerrero es una técnica utilizada para determinar el valor óptimo del parámetro de transformación Box-Cox en series de tiempo. Este método se basa en la minimización de la varianza de los residuos después de aplicar la transformación, lo que ayuda a estabilizar la varianza y mejorar la normalidad de los datos. El método de Guerrero es especialmente útil cuando se trabaja con series de tiempo que presentan heterocedasticidad o variabilidad no constante.

### ¿Cuales son los supuestos de los minimos cuadrados ordinarios?
Los supuestos de los mínimos cuadrados ordinarios (MCO) son los siguientes:
1. Linealidad: La relación entre las variables independientes y la variable dependiente es lineal.
2. Independencia: Las observaciones son independientes entre sí.
3. Homocedasticidad: La varianza de los errores es constante a lo largo de todas las observaciones.
4. Normalidad: Los errores se distribuyen normalmente.
5. No multicolinealidad: Las variables independientes no están altamente correlacionadas entre sí.

### ¿Que es el ARIMA?

ARIMA (AutoRegressive Integrated Moving Average) es un modelo de series de tiempo utilizado para analizar y pronosticar datos que presentan patrones de tendencia y estacionalidad. El modelo ARIMA combina tres componentes: autoregresivo (AR), integrado (I) y media móvil (MA). El componente AR captura la relación entre una observación y un número determinado de observaciones anteriores, el componente I se utiliza para diferenciar los datos para hacerlos estacionarios, y el componente MA captura la relación entre una observación y los errores de las observaciones anteriores. ARIMA es ampliamente utilizado en diversas aplicaciones, como la economía, la meteorología y la ingeniería.

### ¿Que es la bondad de ajuste?

La bondad de ajuste se refiere a qué tan bien un modelo estadístico se ajusta a un conjunto de datos. En el contexto de las series de tiempo, la bondad de ajuste se evalúa utilizando métricas como el coeficiente de determinación (R²), el error cuadrático medio (MSE) o el error absoluto medio (MAE). Estas métricas ayudan a determinar la precisión del modelo en la predicción de los valores futuros y a comparar diferentes modelos para seleccionar el más adecuado para los datos en cuestión. Un buen ajuste indica que el modelo captura adecuadamente las tendencias y patrones presentes en los datos.

### ¿Que explica el R²?

El coeficiente de determinación, comúnmente conocido como R², es una medida estadística que explica la proporción de la variabilidad en la variable dependiente que es explicada por las variables independientes en un modelo de regresión. En otras palabras, R² indica qué tan bien los datos se ajustan al modelo. Un valor de R² cercano a 1 sugiere que el modelo explica una gran parte de la variabilidad de los datos, mientras que un valor cercano a 0 indica que el modelo no explica bien la variabilidad. Es importante tener en cuenta que un alto R² no siempre implica un buen modelo, ya que puede ser influenciado por la cantidad de variables independientes y la complejidad del modelo.
